Using an appropriate Python data structure, implement two programs that demonstrate the behaviour of a stack (LIFO) queue (FIFO). For each program, you should create a menu system to enable the user to manipulate the structure, and the program should keep running until the user chooses to exit. An example menu for a stack would be:

```
1. Push
2. Pop
3. View
```

````python
lifo = list()

last_message = None
while True:
    # Clear the output
    print(chr(27) + "[2J")

    # If there was a message from the previous loop
    if last_message:
        print(f"{last_message=}")
        last_message = None

    # Print the menu
    print("FIFO simulator")
    print(f"{lifo=}")
    print("Menu")
    print("0 = exit")
    print("1 = enqueue")
    print("2 = dequeue")
    
    choice = int(input("What action do you want to take?"))
    
    if choice == 0:
        print("Exiting")
        break
    elif choice == 1:
        # enqueue requires an argument
        item = input("What do you want to enqueue?")
        lifo.append(item)
        last_message = f"enqueued: {item=}"
    elif choice == 2:
        last_message = f"{lifo.pop()=}"
    else:
        last_message = "Invalid choice: {choice}"
````

````python
class FIFO(list):
    _position = 0
    _next_position = 0
    _num_enqueued = 0

    def enqueue(self, item):
        self[self._next_position] = item
        self._next_position += 1
        if self._next_position > len(self):
            self._next_position = 0
        self._num_enqueued += 1

    def deqeue(self):
        if self._position > len(self):
            self._position = 0
        if self._num_enqueued == 0:
            raise ValueError("FIFO is empty")

        item = self[self._position]
        self[self._position] = None
        self._position += 1
        self._num_enqueued -= 1
        return item

    @property
    def length(self):
        return self._num_enqueued


# Create an initial FIFO queue of 5 items
fifo = FIFO()
fifo.extend([None]*5)

last_message = None
while True:
    # Clear the output
    print(chr(27) + "[2J")

    # If there was a message from the previous loop
    if last_message:
        print(f"{last_message=}")
        last_message = None

    # Print the menu
    print("LIFO simulator")
    print(f"{fifo=}")
    print("Menu")
    print("0 = exit")
    print("1 = enqueue")
    print("2 = dequeue")
    
    choice = int(input("What action do you want to take?"))
    
    if choice == 0:
        print("Exiting")
        break
    elif choice == 1:
        # enqueue requires an argument
        item = input("What do you want to enqueue?")
        fifo.enqueue(item)
        last_message = f"enqueued: {item=}"
    elif choice == 2:
        last_message = f"{fifo.deqeue()=}"
    else:
        last_message = "Invalid choice: {choice}"
````

A business has a number of retail outlets across the world. Each outlet keeps a track of its monthly sales and sends a quarterly report to the company's headquarters. Using at least two different data structures, write a program that calculates the average for each outlet and each month, the total for each outlet and each month, and the grand total for the quarter. You should hard code (write data into your program directly) all the data you need once only. The output should be formatted as shown below.


Hint: Remember, a data structure can be nested within another data structure, and the content of one data structure can be used as a reference point in another (see the sample code).

````python
from collections import defaultdict
import numpy


sales_by_month = [
    [None, 'April 18', 'May 18', 'June 18'],
    ['L3', 390, 345, 379],
    ['P2', 250, 270, 300],
    ['N6', 460, 480, 450],
    ['B8', 470, 510, 360],
]

store_lookup = {
    'L3': "London",
    'P2': "Paris",
    'N6': "New York",
    'B8': "Bejing",
}

header_row = sales_by_month[0]

# Dict of the revenue per month for each store
month_tracker = defaultdict(list)
# Dict of the revenue per store for each month
store_tracker = defaultdict(list)

# Get summaries of revenue into list
for store_summary in sales_by_month[1:]:
    store = store_summary[0]
    for i in range(1, len(header_row)):
        month = header_row[i]
        store_revenue = store_summary[i]
        month_tracker[month].append(store_revenue)
        store_tracker[store].append(store_revenue)

print("Quarterly sales report")
print()

grand_total = 0

# Average and sum of revenue of each store per month
for store, quarterly_summary in store_tracker.items():
    store_fullname = store_lookup[store]
    print(f"{store_fullname:<10} {store:>5}:     ${numpy.mean(quarterly_summary):.2f}     ${numpy.sum(quarterly_summary):.2f}")
    grand_total += numpy.sum(quarterly_summary)
print()

# Average and sum of revenue of each month
for month, store_summary in month_tracker.items():
    print(f"{month:<16}:     ${numpy.mean(store_summary):.2f}     ${numpy.sum(store_summary):.2f}")
print()

print(f"Total: ${grand_total}")
````
