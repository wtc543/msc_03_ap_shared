Write a program in Python that asks the user for a sentence and analyses it to establish the number of separate vowels (A, E, I, O, U) and the number of “other” non-space characters contained in the sentence. Your program should output a histogram which indicates a * (star) for each vowel character, followed by a statement indicating the number of other non-space characters.


For example, if the input was: 


The black cat sat up on the orange mat!


The output would be:

```
A: *****
E: ***
I:
O: **
U: *
Other (non-space) Characters: 20
```

````python
input_str = "The black cat sat up on the orange mat!"

def count_vowels(input_str: str) -> str:
    char_counts = {
        "A": 0,
        "E": 0,
        "I": 0,
        "O": 0,
        "U": 0,
        "_": 0,
    }
    for letter in input_str.upper():
        if letter in char_counts.keys():
            char_counts[letter] += 1
        elif letter == " ":
            continue
        else:
            char_counts["_"] += 1
    
    for letter, count in sorted(char_counts.items()):
        if letter == "_":
            print(f"Other (non-space) Characters: {count}")
        else:
            print(f"{letter}: ", end="")
            for _ in range(count):
                print("*", end="")
            else:
                print()


count_vowels(input_str)
````

Write a program in Python that checks if two inputted words are ‘anagrams’. If the user inputs ‘note’ and ‘tone’, the program would output “These are anagrams”. If the user entered 'cat' and 'sat' the program would output "These are not anagrams". Your program should contain a function “isAnagram()” that returns either true or false. The main part of the program should provide a suitable interface for the user. 

````python


def isAnagram(word1: str, word2: str) -> bool:
    if sorted(word1.lower()) == sorted(word2.lower()):
        return True
    else:
        return False
    

print(f"{isAnagram("cat", "sat")=}")
print(f"{isAnagram("tone", "note")=}")

````

````python
class stack_queue(list):
    _position = 0
    _next_position = 0
    _num_enqueued = 0

    def enqueue(self, item):
        self[self._next_position] = item
        self._next_position += 1
        if self._next_position > len(self):
            self._next_position = 0
        self._num_enqueued += 1

    def deqeue(self):
        if self._position > len(self):
            self._position = 0
        if self._num_enqueued == 0:
            raise ValueError("FIFO is empty")

        item = self[self._position]
        self[self._position] = None
        self._position += 1
        self._num_enqueued -= 1
        return item

    @property
    def length(self):
        return self._num_enqueued
````
