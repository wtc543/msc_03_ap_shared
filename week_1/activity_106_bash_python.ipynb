## Question 2

The switch... case structure in Java is often used to build menus in applications. How could you emulate this in Python?

Implement if, elif statement, for example

```python
val = 15

if val == 0:
    print(f"{val=} which is zero")
elif 0 < val <= 10:
    print(f"{val=} which is small")
elif 10 < val <= 20:
    print(f"{val=} which is medium")
elif 20 < val:
    print(f"{val=} which is medium")
else:
    print(f"Invalid value: {val=}")
```

Python does not have a do ..while loop. Consider if this a critical flaw in Python, or something you can work around?

Do while is a while loop that assesses if the while criteria is true at the *end* of the first iteration.


Whilst Python doesn't have this, it is possible to put an else statement at the end of the statement.

```python
val = 0
while val < 10:
    print(f"{val=}")
    val += 1
else:
    print(f"Extra run {val=}")
    val += 1
print(f"Done: {val=}")
```
