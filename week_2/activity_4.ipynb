## Exercise one
Given this plain txt file Download plain txt filecreate regular expressions to do the following:

1. Find the word ‘shrieked’
2. Find the word ‘bleak’
3. Count the number of words that contain ‘pp’
4. Change all the exclamations marks (!) to hash symbols (#)
5. Identify all the words that start with a ‘t” but do not end with an ‘e’, case should be ignored here.


```python
import re

pp_count = 0
with open('activity_4_The_Raven.txt') as f:
    for line in f.readlines():
        if re.search(r'shrieked', line):
            print(f"Found 'shrieked': {line.rstrip()}")
        if re.search(r'bleak', line):
            print(f"Found 'bleak': {line.rstrip()}")
        
        pp_count += len(re.findall(r'pp', line))

        if re.search(r'!', line):
            print(f"before: {line=}")
            line = re.subn(r'!', '#', line)[0]
            print(f"after: {line=}")

        for word in re.findall(r'\w', line):
            if re.search(r'^t.*[^e]$', word, re.IGNORECASE):
                print(f"Word start t, not end in e: {word}")

print(f"Count of string 'pp' {pp_count=}")
```
