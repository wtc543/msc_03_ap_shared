For example in the UnitConvertor exercise (1.16, Activity 6)

```python
class UnitConvertor:
    def __init__(self):
        pass

    def feet_inches_to_meters(self, feet: int, inches: int) -> float:
        metres = feet * 0.3048
        metres += inches * 0.0254
        return metres
```

Could be updated to use assets to make sure that the provided arguments are infact integers. e.g.

```python
class UnitConvertor:
    def __init__(self):
        pass

    def feet_inches_to_meters(self, feet: int, inches: int) -> float:
        assert type(feet) == int, "Non integer provided for feet"
        assert type(inches) == int, "Non integer provided for inches"

        metres = feet * 0.3048
        metres += inches * 0.0254
        return metres
```
