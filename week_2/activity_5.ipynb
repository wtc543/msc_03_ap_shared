## Exercise one

Using either People.xml or People.json (both contain the same data) write a program using the appropriate APIs to translate from one format to another. So you can either use the XML file and translate into JSON or use the JSON file and translate it into XML. You should output the generated format to a file. The outputted file, from your program, can then be compared with the version given here. What, if any, differences are there? Do these differences affect the integrity of the data in the file you generated? If they do, what changes do you need to make your program?



```python
import json
import xml.etree.ElementTree as ET


tree = ET.parse('activity_5_People.xml')
root = tree.getroot()

dict_summary = []

for student in root.findall("./student"):
    student_dict = {}
    for student_attr in student.findall('./'):
        if student_attr.tag == "fullName":
            fullname_dict = {}
            if "title" in student_attr.attrib:
                fullname_dict["title"] = student_attr.attrib['title']
            # print(f"{student_attr.tag=} {student_attr.attrib=}")
            for name_attr in student_attr.findall('./'):
                if name_attr.tag == "other":
                    other_list = []
                    for other in name_attr.findall('./'):
                        other_list.append(other.text)
                        # print(f"{other.tag=} {other.text=}")
                    fullname_dict["other"] = other_list.copy()
                else:
                    if name_attr.attrib:
                        raise ValueError(f"Unhandled attribute '{name_attr.attrib}'")
                    fullname_dict[name_attr.tag] = name_attr.text
                    # print(f"{name_attr.tag} {name_attr.text}")
            student_dict["fullName"] = fullname_dict.copy()
        else:
            if student_attr.attrib:
                raise ValueError(f"Unhandled attribute '{student_attr.attrib}'")
            # print(f"{student_attr.tag=}  {student_attr.text=}")
            student_dict[student_attr.tag] = student_attr.text
    dict_summary.append(student_dict)

with open("activity_5_People_my_version.json", "w", encoding="UTF-8") as f:
    f.write(json.dumps(dict_summary, indent=4))
```
