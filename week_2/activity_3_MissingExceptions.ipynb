```python
num1, num2 = None, None
while num1 is None or num2 is None:
  try:
    num1 = int(input("Enter a number:" ))
    num2 = int(input("Enter another number:" ))
  except ValueError as e:
    print(f"ValueError: {e}")
    print("Try again")

if num1 > num2:
  if num1%num2 == 0:
    print(num1, " is a multiple of ", num2)
  else:
    print(num1, " is NOT a multiple of ", num2)
else:
  if num2%num1 == 0:
    print(num2, " is a multiple of ", num1)
  else:
    print(num2, " is NOT a multiple of ", num1)
```

This is an easy misatke to make by users

```python
randomList = ['67', 53, '3O',72, '10']

for i in randomList:
  try:
    print(int(i) * 10)
  except ValueError:
    i = str(i).replace("O", "0")
    print(int(i) * 10)
```
