## Exercise one

Using the following: PeopleTrainingDate .csv file file,write a Python program to parse (split) the content using string functions only and store it in an appropriate data structure. Print out a table with all the data using the following column headings (which are the first line in the css file): Title,Name,ID,Email,Company,Updated. Remember to consider readability and use formatting instructions in the output. 


```python
file_header = list()
people_training = list()

with open("activity_2_PeopleTrainngDate.csv", encoding="UTF-8") as f:
    # Read the header row
    for item in f.readline().split(','):
        file_header.append(item.rstrip())

    # Read the remaining rows
    for line in f.readlines():
        # Skip empty lines
        if line == "\n":
            continue
        line = line.rstrip()

        # Ignore commas inside quotes
        if '"' in line:
            start_quote = line.find('"')
            end_quote = line.find('"', start_quote + 1) + 1
            l1 = line[:start_quote-1].split(',')
            l1 += [line[start_quote:end_quote],]
            l1 += line[end_quote + 1:].split(',')
        else:
            l1 = line.split('\n')

        # Put all the rows into a list of dictionaries
        row_dict = {}
        for item_no, item in enumerate(l1):
            row_dict[file_header[item_no]] = item.rstrip()
        people_training.append(row_dict)

# Print the header of the table
table_header = "|"
for item in file_header:
    if item in ("Name", "Email", "Company"):
        table_header = f"{table_header} {item:<50} |"
    else:
        table_header = f"{table_header} {item:<20} |"
print(table_header)

# Print each row of the table
for row in people_training:
    table_row = "|"
    for col in file_header:
        if col in ("Name", "Email", "Company"):
            table_row = f"{table_row} {row[col]:<50} |"
        else:
            table_row = f"{table_row} {row[col]:<20} |"
    print(table_row)
```

## Exercise two
Rearrange the data extracted in exercise one and output it to file so that it is sorted in date order (oldest date first) and Updated is the first column in the file (on the left). 



```python
# Create a new key with updated in a sortable format
for row in people_training:
    day, month, year = row['Updated'].split('/')
    row['sort_key'] = f"{year}{month}{day}"

# Write new header and sorted rows to a new file
new_header = ['Updated',] + [c for c in file_header if c != "Updated"]
with open('activity_2_PeopleTrainngDate_sorted.csv', "w", encoding="UTF-8") as f:
    line = ""
    for item in new_header:
        line = line + item + ","
    line = line[:-1]
    f.writelines([line, "\n"])

    for row in sorted(people_training, key=lambda x: x['sort_key']):
        line = ""
        for item in new_header:
            line = line + row[item] + ","
        line = line[:-1]
        f.writelines([line, "\n"])
```

## Exercise three

A recent update to the PeopleTrainingDate file has arrived, PeopleTrainingDateUpdate. Unfortunately, the data has been formatted in a different order. Review the file and write a program to rearrange the data and append it to the file you created in Exercise 2. You will also need to check that there are no errors with the csv format.



```python
update_file_header = ["Updated", "Email", "ID", "Title", "Name", "Company"]
update_people_training = list()

with open('activity_2_PeopleTrainingDateUpdate.csv', encoding='UTF-8') as f:

    # Read the remaining rows
    for line in f.readlines():
        # Skip empty lines
        if line == "\n":
            continue
        line = line.rstrip()

        # Ignore commas inside quotes
        if '"' in line:
            start_quote = line.find('"')
            end_quote = line.find('"', start_quote + 1) + 1
            l1 = line[:start_quote-1].split(',')
            l1 += [line[start_quote:end_quote],]
            l1 += line[end_quote + 1:].split(',')

            # Shift all items in list up 1, apart from first two
            update_pos = len(l1)
            l1.append(l1[update_pos-1])
            update_pos -= 1

            for e in reversed(l1[2:update_pos]):
                l1[update_pos] = e
                update_pos -= 1

            # Replace space in Email field with a ,
            email = l1[1]
            space_position = email.find(" ")
            l1[1] = email[:space_position]
            l1[2] = email[space_position:]
        else:
            l1 = line.split('\n')

        # Put all the rows into a list of dictionaries
        row_dict = {}
        for item_no, item in enumerate(l1):
            row_dict[update_file_header[item_no]] = item.rstrip()
        update_people_training.append(row_dict)

with open('activity_2_PeopleTrainngDate_sorted.csv', "a", encoding="UTF-8") as f:

    for row in update_people_training:
        line = ""
        for item in new_header:
            line = line + row[item] + ","
        line = line[:-1]
        f.writelines([line, "\n"])
```
