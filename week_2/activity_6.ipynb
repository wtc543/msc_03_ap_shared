````python
# Import the JSON document we'll be working with
import json


with open('activity_6_People.json', encoding='UTF-8') as f:
    data = json.load(f)['students']

print(f"{data=}")
````

````python
# Create a data base and collection
import pymongo


client = pymongo.MongoClient("mongodb://localhost:27017")

# Confirm DB doesn't exist
print(f"{client.list_database_names()=}")

# Create a new Database and collection
db = client["university"]
collection = db["studentAdmin"]

# Confirm collections don't exist yet (because it has no documents)
print(f"{db.list_collection_names()=}")

# Confirm database exists
print(f"{client.list_database_names()=}")
````

````python
# Insert first document
ins = collection.insert_one(data[0])

print(f"{ins.inserted_id=}")
print(f"{db.list_collection_names()=}")
print(f"{client.list_database_names()=}")
````

````python
# Insert remaining documents
ins = collection.insert_many(data[1:])

print(f"{len(ins.inserted_ids)=}")
print(f"{ins.inserted_ids=}")
````
